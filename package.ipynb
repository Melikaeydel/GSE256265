```python
import numpy as np
import matplotlib.pyplot as plt

# داده‌های سطح بیان ژن‌ها
gene_names = ['GENE1', 'GENE2', 'GENE3', 'GENE4', 'GENE5']
expression = np.array([2.3, 5.1, 3.8, 4.5, 6.2])

# محاسبه میانگین و انحراف معیار
mean_expr = np.mean(expression)
std_expr = np.std(expression)

print("میانگین سطح بیان ژن‌ها:", round(mean_expr, 2))
print("انحراف معیار:", round(std_expr, 2))

# نمودار میله‌ای برای سطح بیان ژن‌ها
plt.bar(gene_names, expression, color='mediumseagreen')
plt.axhline(mean_expr, color='red', linestyle='--', label='میانگین')
plt.title("سطح بیان ژن‌ها")
plt.ylabel("سطح بیان ژن")
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# داده‌های آموزشی: سطح دو پروتئین
X = [[1.2, 3.4], [2.1, 4.3], [0.8, 2.5], [3.2, 5.1]]
y = [0, 1, 0, 1]  # 0 = سالم، 1 = بیمار

# مدل KNN
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)

# پیش‌بینی وضعیت بیمار جدید
new_sample = [[1.5, 3.7]]
prediction = knn.predict(new_sample)

print("پیش‌بینی وضعیت بیمار جدید:", "بیمار" if prediction[0] == 1 else "سالم")
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# داده‌ها (سطح دو پروتئین برای ۴ فرد)
X = [[1.2, 3.4], [2.1, 4.3], [0.8, 2.5], [3.2, 5.1]]
y = [0, 1, 0, 1]  # 0 = سالم، 1 = بیمار

# مدل KNN
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)

# نمونه جدید
new_sample = [[1.5, 3.7]]
prediction = knn.predict(new_sample)

# گراف
for i in range(len(X)):
    color = 'green' if y[i] == 0 else 'red'
    label = 'سالم' if y[i] == 0 else 'بیمار'
    plt.scatter(X[i][0], X[i][1], c=color, label=label if i < 2 else "", s=100, edgecolors='k')

# بیمار جدید
plt.scatter(new_sample[0][0], new_sample[0][1], c='blue', label='بیمار جدید', s=150, marker='*', edgecolors='black')

plt.xlabel("سطح پروتئین A")
plt.ylabel("سطح پروتئین B")
plt.title("مدل ساده KNN برای تشخیص بیماری")
plt.legend()
plt.grid(True)
plt.show()

print("پیش‌بینی مدل:", "بیمار" if prediction[0] == 1 else "سالم")
```
