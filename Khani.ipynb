```python
# Potential Energy
!gmx energy -f em.edr -o potential.xvg
```

```python
import matplotlib.pyplot as plt

time = []
potential = []

with open("potential.xvg") as f:
    for line in f:
        if line.startswith(('#', '@')):  # Skip comments/headers
            continue
        parts = line.split()
        if len(parts) >= 2:
            time.append(float(parts[0]))
            potential.append(float(parts[1]))

plt.figure(figsize=(8,5))
plt.plot(time, potential, label="Potential Energy", color="blue", lw=1.5)
plt.xlabel("Step")
plt.ylabel("Potential Energy (kJ/mol)")
plt.title("Potential Energy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
!gmx rms -s em.tpr -f md_fitted.xtc -o rmsd_xtal.xvg -tu ns
```

```python
import matplotlib.pyplot as plt

time = []
rmsd = []

with open("rmsd_xtal.xvg") as f:
    for line in f:
        if line.startswith(('#', '@')):  # Skip headers and comments
            continue
        parts = line.split()
        if len(parts) >= 2:
            time.append(float(parts[0]))
            rmsd.append(float(parts[1]))

plt.figure(figsize=(8,5))
plt.plot(time, rmsd, label="RMSD (Backbone)", color="green", lw=1.5)
plt.xlabel("Time (ns)")
plt.ylabel("RMSD (nm)")
plt.title("RMSD vs Time (NPT/MD Trajectory)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
!gmx rmsf -s md_0_2.tpr -f md_fitted.xtc -o rmsf_ca.xvg -res
```

```python
import matplotlib.pyplot as plt

residues = []
rmsf = []

with open("rmsf_ca.xvg") as f:
    for line in f:
        if line.startswith(('#', '@')):  # Skip comments/headers
            continue
        parts = line.split()
        if len(parts) >= 2:
            residues.append(int(float(parts[0])))  # Residue number
            rmsf.append(float(parts[1]))          # RMSF value

plt.figure(figsize=(10,5))
plt.bar(residues, rmsf, color="orange", width=0.8)
plt.xlabel("Residue Number")
plt.ylabel("RMSF (nm)")
plt.title("Residue-wise RMSF (Backbone)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
