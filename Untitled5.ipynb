```python
import GEOparse

# دانلود داده‌های GSE256265 (ممکن است چند دقیقه طول بکشد)
gse = GEOparse.get_GEO(geo="GSE256265", destdir="./")

# نمایش اطلاعات کلی
print("شناسه پلتفرم:", gse.metadata['platform_id'][0])
print("تعداد نمونه‌ها:", len(gse.gsms))
print("تعداد ژن‌ها/پروب‌ها:", gse.gpls[list(gse.gpls.keys())[0]].table.shape[0])
```

```python
# انتخاب اولین نمونه
sample_id = list(gse.gsms.keys())[0]
sample_data = gse.gsms[sample_id]

# نمایش متادیتای نمونه
print("متادیتای نمونه:", sample_data.metadata)

# نمایش ۵ ردیف اول داده‌های بیان
print("داده‌های بیان:", sample_data.table.head())
```

```python
platform_id = list(gse.gpls.keys())[0]
platform_data = gse.gpls[platform_id]

# نمایش آنوتاسیون ژن‌ها (ستون‌های مهم: ID_REF, Gene Symbol, ENTREZ_GENE_ID)
print(platform_data.table.head())
```

```python
import pandas as pd

# ساخت لیست داده‌های بیان همه نمونه‌ها
expression_data = []
for sample_id, sample in gse.gsms.items():
    expression_data.append(sample.table['VALUE'])  # ستون VALUE حاوی مقادیر بیان است

# تبدیل به DataFrame
df_expression = pd.DataFrame(expression_data, index=gse.gsms.keys()).T  # T برای جابجایی سطر/ستون

# نمایش ابعاد داده
print("ابعاد ماتریس بیان (ژن‌ها × نمونه‌ها):", df_expression.shape)
df_expression.head()
```

```python
print(df_expression.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df_expression.mean().plot.hist(bins=50, color='skyblue')
plt.title("توزیع میانگین بیان ژن‌ها در نمونه‌ها")
plt.xlabel("میانگین بیان")
plt.ylabel("تعداد نمونه‌ها")
plt.show()
```

```python
import seaborn as sns

# محاسبه ماتریس همبستگی
corr_matrix = df_expression.corr()

# رسم هیت‌مپ
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("همبستگی بین نمونه‌ها")
plt.show()
```

```python
import seaborn as sns

# انتخاب 50 ژن با بیشترین واریانس (برای سادگی)
top_genes = df_expression.var(axis=1).nlargest(50).index
df_top = df_expression.loc[top_genes]

# رسم هیت‌مپ
plt.figure(figsize=(12, 8))
sns.clustermap(df_top.corr(), cmap="viridis", figsize=(10, 10))
plt.title("همبستگی بین نمونه‌ها")
plt.show()
```
