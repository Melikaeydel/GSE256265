##instal package and show data

```python
import pandas as pd

# بارگذاری داده‌های بیان ژن
expr_colo = pd.read_csv('esets_colorectal_exprs_genes.txt', sep='\t', index_col=0)

# نمایش ابعاد داده
print("Shape of colorectal expression matrix:", expr_colo.shape)

# نمایش چند سطر اول
expr_colo.head()
```

```python
import pandas as pd

# بارگذاری فایل SDRF مربوط به روده
sdrf_colo = pd.read_csv('colo.txt', sep='\t')

# استخراج نمونه و وضعیت بیماری
sample_info_colo = sdrf_colo[['Source Name', 'Factor Value[disease]']]
sample_info_colo.columns = ['sample', 'group']

# اصلاح نام نمونه‌ها برای تطابق با نام ستون‌های expression matrix
sample_info_colo['sample'] = sample_info_colo['sample'].str.strip()

# نمایش گروه‌ها
print(sample_info_colo['group'].value_counts())
sample_info_colo.head()
```

```python
print(sample_info_colo['sample'].unique()[:10])
```

##seprate tumor and control

```python
# ساخت دیکشنری نمونه -> گروه بیماری
sample_group_dict = dict(zip(sample_info_colo['sample'], sample_info_colo['group']))

# گرفتن نمونه‌های مشترک بین expression matrix و فایل SDRF
common_samples = [s for s in expr_colo.columns if s in sample_group_dict]

expr_colo_filtered = expr_colo[common_samples]

# جدا کردن نمونه‌های تومور و سالم
tumor_samples = [s for s in common_samples if sample_group_dict[s] == 'tumor']
normal_samples = [s for s in common_samples if sample_group_dict[s] == 'normal']

print(f"Number of tumor samples: {len(tumor_samples)}")
print(f"Number of normal samples: {len(normal_samples)}")
```

##logFc

```python
# میانگین بیان ژن در گروه‌ها
mean_expr_tumor = expr_colo_filtered[tumor_samples].mean(axis=1)
mean_expr_normal = expr_colo_filtered[normal_samples].mean(axis=1)

# محاسبه تفاوت بیان (logFC = میانگین تومور - میانگین نرمال)
logFC = mean_expr_tumor - mean_expr_normal

# مرتب سازی ژن‌ها بر اساس logFC نزولی
top_genes = logFC.sort_values(ascending=False).head(100)

print(top_genes)
```

##PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# نرمال‌سازی ساده: log2(x+1)
expr_colo_log = expr_colo_filtered.applymap(lambda x: np.log2(x + 1))

# PCA با 2 کامپوننت
pca = PCA(n_components=2)
pca_result = pca.fit_transform(expr_colo_log.T)  # ترانهاده چون نمونه‌ها ستون‌ها هستن

# دیتافریم PCA با گروه بیماری
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'], index=expr_colo_log.columns)
pca_df['group'] = pca_df.index.map(sample_group_dict)

# رسم
plt.figure(figsize=(8,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='group', palette=['green', 'red'])
plt.title('PCA plot of colorectal samples')
plt.show()
```

##Volcanoplot

```python
from scipy.stats import ttest_ind
import numpy as np

# t-test بین گروه‌ها برای هر ژن
p_values = []
for gene in expr_colo_log.index:
    group1 = expr_colo_log.loc[gene, tumor_samples]
    group2 = expr_colo_log.loc[gene, normal_samples]
    stat, p = ttest_ind(group1, group2, equal_var=False)
    p_values.append(p)

p_values = np.array(p_values)
logFC = mean_expr_tumor - mean_expr_normal

# ساخت DataFrame برای ولکانولوپ
volcano_df = pd.DataFrame({
    'gene': expr_colo_log.index,
    'logFC': logFC,
    'pval': p_values
})

# محاسبه -log10(p-value)
volcano_df['neg_log10_pval'] = -np.log10(volcano_df['pval'])

# رسم ولکانولوپ
plt.figure(figsize=(10,7))
sns.scatterplot(data=volcano_df, x='logFC', y='neg_log10_pval', alpha=0.5)

# خطوط آستانه برای logFC و p-value
plt.axvline(x=1, color='red', linestyle='--')
plt.axvline(x=-1, color='red', linestyle='--')
plt.axhline(y=-np.log10(0.05), color='blue', linestyle='--')

plt.xlabel('log2 Fold Change')
plt.ylabel('-log10 p-value')
plt.title('Volcano plot of colorectal cancer DEGs')
plt.show()
```

##heatmap

```python
import seaborn as sns

# انتخاب 50 ژن بالا و پایین بر اساس logFC
top_up_genes = volcano_df[volcano_df['logFC'] > 1].sort_values(by='logFC', ascending=False).head(50)['gene']
top_down_genes = volcano_df[volcano_df['logFC'] < -1].sort_values(by='logFC').head(50)['gene']
selected_genes = list(top_up_genes) + list(top_down_genes)

# دیتافریم بیان ژن‌ها
expr_selected = expr_colo_log.loc[selected_genes, common_samples]

# نرمال‌سازی برای هیت‌مپ (Z-score)
expr_selected_z = expr_selected.apply(lambda x: (x - x.mean()) / x.std(), axis=1)

# رسم هیت‌مپ
plt.figure(figsize=(12,10))
sns.clustermap(expr_selected_z, cmap='vlag', col_cluster=True, row_cluster=True)
plt.title('Heatmap of top DE genes')
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# 1. بارگذاری داده‌ها (اگر هنوز بارگذاری نکردی)
expr_colorectal = pd.read_csv('/content/esets_colorectal_exprs_genes.txt', sep='\t', index_col=0)
sdrf_colorectal = pd.read_csv('/content/colo.txt', sep='\t', index_col=0)

# 2. استخراج نمونه‌های تومور و نرمال
tumor_samples = sdrf_colorectal[sdrf_colorectal['Factor Value[disease]'].str.lower() == 'tumor'].index.tolist()
normal_samples = sdrf_colorectal[sdrf_colorectal['Factor Value[disease]'].str.lower() == 'normal'].index.tolist()

# 3. فیلتر کردن نمونه‌هایی که در داده بیان وجود دارند
tumor_samples = [s for s in tumor_samples if s in expr_colorectal.columns]
normal_samples = [s for s in normal_samples if s in expr_colorectal.columns]

print(f"Number of tumor samples: {len(tumor_samples)}")
print(f"Number of normal samples: {len(normal_samples)}")

# 4. محاسبه میانگین بیان و t-test برای هر ژن
log2fc_list = []
pval_list = []

for gene in expr_colorectal.index:
    expr_tumor = expr_colorectal.loc[gene, tumor_samples]
    expr_normal = expr_colorectal.loc[gene, normal_samples]

    mean_tumor = np.mean(expr_tumor)
    mean_normal = np.mean(expr_normal)

    # محاسبه log2 fold change
    log2fc = mean_tumor - mean_normal

    # t-test
    t_stat, p_val = ttest_ind(expr_tumor, expr_normal, equal_var=False, nan_policy='omit')

    log2fc_list.append(log2fc)
    pval_list.append(p_val)

# 5. ساخت دیتافریم DEG
df_deg = pd.DataFrame({
    'log2FC': log2fc_list,
    'p_value': pval_list
}, index=expr_colorectal.index)

# 6. اضافه کردن -log10p برای رسم ولکانو پلات
df_deg['-log10p'] = -np.log10(df_deg['p_value'])

# 7. نمایش 10 ژن برتر بر اساس log2FC
top10_genes = df_deg.sort_values('log2FC', ascending=False).head(10)
print("Top 10 upregulated genes in colorectal tumor:")
print(top10_genes)

```

```python
import pandas as pd
from scipy.stats import ttest_ind
import numpy as np

# 1. نمونه‌های تومور و نرمال رو از متادیتا جدا کن
tumor_samples_colorectal = sdrf_colorectal[sdrf_colorectal['Factor Value[disease]'].str.lower() == 'tumor'].index.tolist()
normal_samples_colorectal = sdrf_colorectal[sdrf_colorectal['Factor Value[disease]'].str.lower() == 'normal'].index.tolist()

# 2. از expression دیتافریم ستون‌های مربوط به نمونه‌ها رو بگیر
expr_tumor = expr_colorectal[tumor_samples_colorectal]
expr_normal = expr_colorectal[normal_samples_colorectal]

# 3. برای هر ژن t-test بگیر
results = []
for gene in expr_colorectal.index:
    group1 = expr_tumor.loc[gene].values
    group2 = expr_normal.loc[gene].values
    stat, p = ttest_ind(group1, group2, equal_var=False)
    mean_diff = np.mean(group1) - np.mean(group2)
    results.append([gene, mean_diff, p])

# 4. تبدیل به دیتافریم و اضافه کردن ستون‌های log2FC و p-value
df_deg_colorectal = pd.DataFrame(results, columns=['gene', 'log2FC', 'p_value']).set_index('gene')

# 5. محاسبه -log10p برای رسم ولکان
df_deg_colorectal['-log10p'] = -np.log10(df_deg_colorectal['p_value'])

# 6. گرفتن 10 ژن برتر بر اساس log2FC
top10_colorectal = df_deg_colorectal.sort_values('log2FC', ascending=False).head(10)

print("Top 10 colorectal genes:")
print(top10_colorectal)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind

# فرض: expr_tumor_lung و expr_normal_lung آماده‌ان (هر دو DataFrame با ژن‌ها در ردیف‌ها و نمونه‌ها در ستون‌ها)

# میانگین و انحراف معیار هر ژن در دو گروه
mean_tumor = expr_tumor_lung.mean(axis=1)
mean_normal = expr_normal_lung.mean(axis=1)

# محاسبه log2FC
log2FC = np.log2(mean_tumor + 1e-9) - np.log2(mean_normal + 1e-9)  # +1e-9 برای جلوگیری از صفر

# محاسبه p-value با t-test دو نمونه‌ای
p_values = []
for gene in expr_tumor_lung.index:
    t_stat, p_val = ttest_ind(expr_tumor_lung.loc[gene], expr_normal_lung.loc[gene], equal_var=False)
    p_values.append(p_val)

# ساخت دیتافریم نتایج DEG
df_deg_lung = pd.DataFrame({
    'log2FC': log2FC,
    'p_value': p_values
})

# محاسبه -log10(p-value) برای رسم
df_deg_lung['-log10p'] = -np.log10(df_deg_lung['p_value'] + 1e-300)

# مرتب‌سازی بر اساس مقدار مطلق log2FC و p-value (مثلاً اول p-value کوچک‌تر، بعد log2FC بزرگ‌تر)
df_deg_lung = df_deg_lung.sort_values(by=['p_value', 'log2FC'])

# نمایش 10 ژن برتر
print(df_deg_lung.head(10))
```

```python
# لیست ژن‌ها از دو سری داده
top10_colorectal = ['FOXQ1', 'MMP1', 'CEMIP', 'KRT23', 'MMP3', 'INHBA', 'TACSTD2', 'CRNDE', 'MMP7', 'CXCL8']
top10_lung = ['SFTPC', 'IGF2BP3', 'COL11A1', 'SLC2A1', 'AQP4', 'EXO1', 'LAMP3', 'LRRK2', 'SFTPD', 'IGSF9']

# محاسبه اشتراک
common_genes = set(top10_colorectal).intersection(set(top10_lung))

print("🔗 ژن‌های مشترک بین روده و ریه (Top 10):")
print(common_genes if common_genes else "هیچ ژن مشترکی وجود ندارد.")
```

```python
# انتخاب 50 ژن برتر از هر دیتافریم
top50_lung_genes = df_deg_lung.sort_values('log2FC', ascending=False).head(50).index
top50_colorectal_genes = df_deg_colorectal.sort_values('log2FC', ascending=False).head(50).index

# پیدا کردن ژن‌های مشترک
common_top_genes = set(top50_lung_genes).intersection(set(top50_colorectal_genes))

# نمایش
print("✅ ژن‌های مشترک بین 50 ژن برتر روده و ریه:")
print(common_top_genes)
```
