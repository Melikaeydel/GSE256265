```python
import GEOparse

gse = GEOparse.get_GEO("GSE256265", destdir=".")
print(gse)
```

```python
# لیست کردن GSM ها (نمونه‌ها)
print(gse.gsms.keys())
```

```python
sample = gse.gsms["GSM8092222"]
print(sample.metadata)  # اطلاعات متادیتای اون نمونه
```

```python
# انتخاب یکی از نمونه‌ها، مثلاً اولین نمونه
first_sample = list(gse.gsms.values())[0]

# گرفتن داده‌های expression به صورت DataFrame
df = first_sample.table

# نمایش ۵ ردیف اول
df.head()
```

```python
import matplotlib.pyplot as plt

# رسم هیستوگرام از مقادیر بیان
df['VALUE'].hist(bins=50)
plt.title("Distribution of Gene Expression")
plt.xlabel("Expression Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import pandas as pd

# استخراج بیان ژن از چند نمونه و ترکیب آن‌ها در یک DataFrame
samples = list(gse.gsms.keys())[:3]  # فقط ۳ نمونه اول

all_data = pd.DataFrame()

for gsm_id in samples:
    sample = gse.gsms[gsm_id]
    df = sample.table[["ID_REF", "VALUE"]].copy()
    df.columns = ["ID_REF", gsm_id]

    if all_data.empty:
        all_data = df
    else:
        all_data = pd.merge(all_data, df, on="ID_REF")

all_data.head()
```

```python
# رسم نمودار boxplot برای مقایسه بیان در نمونه‌های مختلف
all_data.drop("ID_REF", axis=1).boxplot()
plt.title("Comparison of Gene Expression Across Samples")
plt.ylabel("Expression Level")
plt.xticks(rotation=45)
plt.show()
```
