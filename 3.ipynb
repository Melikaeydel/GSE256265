```python
import pandas as pd

# مسیر فایل در Colab
file_path = '/content/GSE256265_series_matrix (1).txt'

# خواندن فایل با pandas
df = pd.read_csv(file_path, sep="\t", comment="!", header=0)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ترسیم نمودار Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Boxplot of Gene Expression")
plt.xticks(rotation=90)
plt.show()
```

```python
# انتخاب فقط ستون‌های عددی از DataFrame
df_numeric = df.select_dtypes(include=['number'])

# نمایش پنج سطر اول داده‌های عددی
print(df_numeric.head())
```

```python
# محاسبه میانگین هر ستون عددی
means = df_numeric.mean(axis=0)
print(means)
```

```python
# ترسیم نمودار Boxplot برای داده‌های عددی
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_numeric)
plt.title("Boxplot of Gene Expression")
plt.xticks(rotation=90)
plt.show()

# ترسیم Heatmap برای ماتریس همبستگی داده‌های عددی
plt.figure(figsize=(12, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# دسترسی به داده‌های نمونه خاص (مثلاً نمونه 'GSM8092189')
sample_data = df['GSM8092189']
print(sample_data.head())
```

```python
# رسم هیستوگرام برای یک نمونه
plt.figure(figsize=(8, 5))
df['GSM8092189'].astype(float).hist(bins=50)
plt.title("Histogram of Sample GSM8092189")
plt.xlabel("Expression Level")
plt.ylabel("Frequency")
plt.show()
```

```python
# رسم boxplot برای چند نمونه
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['GSM8092189', 'GSM8092190', 'GSM8092191']].astype(float))
plt.title("Boxplot of Selected Samples")
plt.xlabel("Samples")
plt.ylabel("Expression Level")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df[['GSM8092189', 'GSM8092190', 'GSM8092191']].astype(float))
plt.title("Violin Plot of Samples")
plt.show()
```

```python
# رسم نمودار خطی برای یک ژن خاص (مثلاً اولین ردیف)
gene_values = df.iloc[0, 1:].astype(float)  # اولین ژن
plt.plot(gene_values.values)
plt.title("Expression of Gene 1 Across Samples")
plt.xlabel("Sample Index")
plt.ylabel("Expression Level")
plt.show()
```
